### 2. **Section : Pourquoi un système de recommandation ? (Data Story)**
#### **Objectif : Expliquer vos motivations et fournir une analyse exploratoire des données.**

- **Introduction générale :**
  - Imagine un monde où chaque utilisateur est confronté à une bibliothèque infinie de films, un véritable labyrinthe de choix. Comment décider ? Grâce à nos systèmes de recommandation, nous voulons non seulement simplifier cette tâche, mais aussi rendre chaque expérience unique et personnalisée. Pour comprendre pourquoi de tels systèmes sont essentiels, explorons les données et ce qu’elles révèlent sur les films, leurs genres, et les préférences des spectateurs.

- **Sections :**
  1. **L’ampleur des données disponibles :** (ou alors "Un monde saturé de choix.")
     - **Visu :**  
       - Un histogramme ou graphique montrant la distribution du nombre de films produits par année (avec les proportions par pays). 
         - *Ce graphique met en évidence l’explosion du nombre de films produits au fil des années, notamment avec l’essor des plateformes de streaming, du monde hollywoodien et des contenus indépendants. Les proportions par pays permettent de voir les tendances de production dans différentes régions du monde.*
       - Map interactive du nombre de films produits par pays en fonction des années. 
         - *Ici nous observons l’évolution de la production de films dans le monde. Nous y voyons les pays les plus prolifiques. Il souligne la diversité des contenus cinématographiques et les tendances régionales. Nous voyons que le cinéma était initialement destiné aux pays riches et occidentaux et qu'il s'est démocratisé partout.* (peut etre mettre des éléments historiques pour expliquer les tendances, et de comparer avec les tendances actuelles)
       - Un graphique en barres montrant la distribution des genres de films par pays aujourd'hui.
         - *Ce graphique permet de visualiser la production de films par pays et de mettre en évidence les genres les plus populaires dans chaque région. Peut-on alors trouver des points communs entre certaines ethinicités? Est-ce que les gens aiment des films particuliers? * (-> lien vers la seconde partie , peut etre relever des tendances particulières)
        
     - Insight : Souligner que le volume croissant de contenu rend la découverte difficile pour les utilisateurs.
  2. **L’importance des préférences utilisateurs :**
     - **Visu :**
       - Une heatmap montrant la corrélation entre genres populaires et ratings.  
         - *Cette visualisation montre les genres qui reçoivent généralement de bonnes évaluations et ceux qui sont plus polarisants. Les genres comme la romance ou la science-fiction peuvent avoir des évaluations très variables selon les spectateurs, reflétant des goûts spécifiques et divisés.*
       - Un graphique en barres des genres les plus populaires.
         - *Ce graphique met en lumière les genres qui attirent le plus grand nombre de spectateurs. Par exemple, les comédies et les drames sont souvent des choix universels, tandis que les genres de niche comme le documentaire peuvent avoir une audience plus restreinte. Une compréhension des genres populaires est essentielle pour développer un système qui capte l’intérêt d’un large public tout en satisfaisant des préférences plus uniques.*
       - Un graphe en radar comparant les préférences des utilisateurs entre genres (par ex. Action, Romance, Science-Fiction).  
         - *Ce graphe montre comment les préférences pour certains genres (ex. action, romance, science-fiction) diffèrent d’un groupe d’utilisateurs à l’autre. Ce résultat souligne que chaque spectateur a une "empreinte de goût" unique, ce qui justifie l’importance de recommandations personnalisées.*
     - Insight : Montrer que les goûts diffèrent selon les utilisateurs et qu’un système généralisé ne suffit pas.

  3. **Motivations pour un système de recommandation :**
     - **Visu :**
       - Une carte interactive affichant les films populaires dans différents pays (ex. genres ou notes préférés par région).  
         - *Cette carte met en évidence que les goûts en matière de films varient considérablement selon les régions. Par exemple, les comédies romantiques peuvent être très populaires en Europe, tandis que l’action domine en Asie. Les systèmes de recommandation doivent s’adapter aux préférences régionales pour rester pertinents.*
       - Un diagramme en barres des notes moyennes des utilisateurs par catégorie de film.
         - *Ce graphique compare les évaluations moyennes pour différents genres de films, montrant par exemple que les documentaires et les drames reçoivent souvent des notes élevées, tandis que les comédies peuvent polariser davantage. Certains genres sont intrinsèquement mieux évalués, ce qui peut influencer les recommandations faites aux spectateurs.*
       - Un nuage de points des notes moyennes en fonction du nombre de notes par film. (montrer la popularité des films et leur distribution)
         - *Ce graphique illustre que certains films (souvent très populaires) accumulent des milliers de notes, tandis que d’autres restent presque inconnus avec très peu d’évaluations. La popularité et la qualité perçue ne sont pas toujours corrélées. Par exemple, certains films très notés peuvent être moins populaires, et vice versa.*
       - Corrélation entre les notes des utilisateurs et les genres de films ou le nombre de notes, ou la popularité des films, etc. 
         - *Les corrélations révèlent des patterns intéressants, comme le fait que les films populaires (beaucoup de notes) tendent à recevoir des évaluations modérées, tandis que les films moins connus (peu de notes) ont des scores plus extrêmes (soit très bas, soit très hauts). Cela souligne l’importance de diversifier les recommandations pour permettre à l’utilisateur de découvrir des films moins populaires mais bien notés, au lieu de toujours promouvoir les blockbusters.*
     - Insight : Souligner que les systèmes personnalisés facilitent la découverte de contenu.


- **Conclusion :**
- *Ces visualisations montrent que l’univers des films est un espace vaste et complexe. Les préférences individuelles, les variations régionales et la saturation du contenu rendent la découverte de films adaptés aux goûts de chacun difficile sans une intervention intelligente. C’est ici que nos systèmes de recommandation entrent en jeu, en utilisant ces observations pour simplifier et enrichir votre expérience cinématographique. Prêts à plonger dans le monde des recommandations personnalisées ?*

### 3. **Section : Exploration des systèmes de recommandation**
#### **Objectif : Présenter et démontrer les trois systèmes.**

- **Introduction générale :**
  - Brève explication des trois types de systèmes créés.  

- **Sous-sections pour chaque système :**
  1. **Système basé sur la distance (Distance-based Recommender) :**
     - **Explication** : Calcul des similarités entre utilisateurs ou films à l’aide de métriques comme la distance cosinus ou Euclidienne.
     - **Interface interactive :**
       - Entrer un utilisateur fictif ou un film pour obtenir des recommandations basées sur des préférences similaires.  
     - **Visu :**
       - Graphe des similarités entre les utilisateurs (par ex. clustering d’utilisateurs proches dans un espace 2D).  
       - Carte des films similaires pour un film donné (ex. visualisation des genres les plus proches).
       - Comparaison des notes entre utilisateurs similaires pour un film donné.
       - Heatmap des notes des utilisateurs pour les films recommandés.
       - Afficher les recommandations pour un utilisateur en fonction des notes des utilisateurs similaires.

  2. **Système basé sur la SVD (SVD-based Recommender) :**
     - **Explication** : Réduction dimensionnelle des matrices utilisateur-film pour prédire les notes manquantes.
     - **Interface interactive :**
       - Afficher les recommandations pour un utilisateur en fonction des prédictions SVD.  
     - **Visualisation suggérée :**
       - Graphe montrant la décomposition en valeurs singulières (distribution des singularités).
       - Comparaison entre les notes réelles et prédites pour certains films.

  3. **Système de factorisation matricielle (Matrix Factorization Recommender) :**
     - **Explication** : Factorisation des matrices utilisateur-film pour découvrir des relations latentes.
     - **Interface interactive :**
       - Entrer un utilisateur ou un film pour afficher les recommandations basées sur les facteurs latents.  
     - **Visualisation suggérée :**
       - Représentation 2D des vecteurs utilisateur et film.  
       - Graphe montrant la convergence de l’algorithme (erreur en fonction des itérations? Je sais pas si c'est relevant).
