- **Sous-sections pour chaque système :**
  1. **Système basé sur la distance (Distance-based Recommender) :**
     - **Explication** : Calcul des similarités entre utilisateurs ou films à l’aide de métriques comme la distance cosinus ou Euclidienne.
     - **Interface interactive :**
       - Entrer un utilisateur fictif ou un film pour obtenir des recommandations basées sur des préférences similaires.  
     - **Visu :**
       - Graphe des similarités entre les utilisateurs (par ex. clustering d’utilisateurs proches dans un espace 2D).  
       - Carte des films similaires pour un film donné (ex. visualisation des genres les plus proches).
       - Comparaison des notes entre utilisateurs similaires pour un film donné.
       - Heatmap des notes des utilisateurs pour les films recommandés.
       - Afficher les recommandations pour un utilisateur en fonction des notes des utilisateurs similaires.

  2. **Système basé sur la SVD (SVD-based Recommender) :**
     - **Explication** : Réduction dimensionnelle des matrices utilisateur-film pour prédire les notes manquantes.
     - **Interface interactive :**
       - Afficher les recommandations pour un utilisateur en fonction des prédictions SVD.  
     - **Visualisation suggérée :**
       - Graphe montrant la décomposition en valeurs singulières (distribution des singularités).
       - Comparaison entre les notes réelles et prédites pour certains films.

  3. **Système de factorisation matricielle (Matrix Factorization Recommender) :**
     - **Explication** : Factorisation des matrices utilisateur-film pour découvrir des relations latentes.
     - **Interface interactive :**
       - Entrer un utilisateur ou un film pour afficher les recommandations basées sur les facteurs latents.  
     - **Visualisation suggérée :**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import widgets


# df_final_dataset['release_date_clean'] = pd.to_datetime(df_final_dataset['release_date_clean'])

# Préparer les données par pays et par année
def prepare_country_data(df):
    # Extraire les pays de la colonne 'countries'
    df_countries = (
        df['countries']
        .str.split(', ', expand=True)
        .stack()
        .reset_index(level=1, drop=True)
        .rename("country")
    )
    df_with_countries = df[['release_date_clean']].join(df_countries)
    df_with_countries['year'] = df_with_countries['release_date_clean'].dt.year
    
    # Nombre total de films par année
    total_by_year = df_with_countries['year'].value_counts().sort_index()
    
    # Nombre de films par pays et par année
    country_year_counts = (
        df_with_countries.groupby(['year', 'country'])
        .size()
        .unstack(fill_value=0)
    )
    
    # Trier les pays par leur nombre total de films (du plus grand au plus petit)
    country_totals = country_year_counts.sum(axis=0).sort_values(ascending=False)
    
    return total_by_year, country_year_counts, country_totals

# Tracer le graphique interactif
def plot_movies_by_year(total_by_year, country_year_counts, selected_country):
    plt.figure(figsize=(12, 6))
    
    # Histogramme total des films
    plt.bar(total_by_year.index, total_by_year.values, label="Total des films par année", alpha=0.5, color='blue')
    
    # Histogramme des films par pays
    if selected_country in country_year_counts.columns:
        plt.bar(country_year_counts.index, country_year_counts[selected_country], 
                label=f"Films de {selected_country}", alpha=0.7, color='orange')
    
    # Ajouter des légendes et des étiquettes
    plt.xlabel('Année')
    plt.ylabel('Nombre de films')
    plt.title(f"Nombre de films par année : Total vs {selected_country}")
    plt.legend()
    plt.show()

# Charger et préparer les données
total_by_year, country_year_counts, country_totals = prepare_country_data(df_final_dataset)

# Créer un widget pour sélectionner les pays
@interact(country=widgets.Dropdown(options=country_totals.index, description='Pays'))
def interactive_plot(country):
    plot_movies_by_year(total_by_year, country_year_counts, country)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Préparation des données
def prepare_genre_country_data(df):
    # Extraire les genres et les pays
    df_genres = (
        df['genres_cmu']
        .str.split(', ', expand=True)
        .stack()
        .reset_index(level=1, drop=True)
        .rename("genre")
    )
    df_countries = (
        df['countries']
        .str.split(', ', expand=True)
        .stack()
        .reset_index(level=1, drop=True)
        .rename("country")
    )
    
    # Fusionner les deux ensembles de données
    df_with_genres_countries = df[['genres_cmu', 'countries']].join(df_genres).join(df_countries)
    
    # Calculer le nombre de films par genre et par pays
    genre_country_counts = (
        df_with_genres_countries.groupby(['country', 'genre'])
        .size()
        .unstack(fill_value=0)
    )
    
    return genre_country_counts

# Charger et préparer les données
genre_country_counts = prepare_genre_country_data(df_final_dataset)

# Trier les pays en fonction du total des films produits
genre_country_counts['Total'] = genre_country_counts.sum(axis=1)
genre_country_counts = genre_country_counts.sort_values('Total', ascending=False)
countries = genre_country_counts.index.tolist()

# Supprimer la colonne 'Total' après tri
genre_country_counts = genre_country_counts.drop(columns=['Total'])

def plot_top_genres_by_country(country):
    clear_output(wait=True)  # Nettoyer la sortie avant de tracer un nouveau graphique
    
    if country not in genre_country_counts.index:
        print(f"Le pays '{country}' n'existe pas dans les données.")
        return
    
    # Obtenir les genres pour le pays sélectionné
    country_data = genre_country_counts.loc[country]
    
    # Trier les genres par nombre d'occurrences et regrouper les autres
    top_genres = country_data.nlargest(7)
    other_count = country_data.sum() - top_genres.sum()
    genre_counts = pd.concat([top_genres, pd.Series({'Other': other_count})])
    
    # Tracer le graphique
    plt.figure(figsize=(10, 6))
    genre_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Distribution des genres pour {country}', fontsize=16)
    plt.xlabel('Genres', fontsize=12)
    plt.ylabel('Nombre de films', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Widget pour sélectionner un pays
country_selector = widgets.Dropdown(
    options=countries,
    description='Pays:',
    value=countries[0],  # Premier pays par défaut
    style={'description_width': 'initial'},
)

# Fonction pour mettre à jour le graphique lors de la sélection d'un pays
def update_graph(change):
    plot_top_genres_by_country(change.new)

# Associer la fonction au widget
country_selector.observe(update_graph, names='value')

# Afficher le widget et initialiser le graphique
display(country_selector)
plot_top_genres_by_country(country_selector.value)

```

```python
import pandas as pd
import ipywidgets as widgets
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
df_final_dataset = df_final_dataset_.copy()

# Exemple de données (remplacer par vos vraies données)
# df_final_dataset = pd.read_csv("your_file.csv")  # Charger votre dataset ici
# Assurez-vous que la colonne 'genres_cmu' contient des genres et que 'countries' contient les pays

# Nettoyage des NaN dans les colonnes concernées
df_final_dataset['genres_cmu'].dropna(inplace=True)
df_final_dataset['countries'].dropna(inplace=True)
# genres en string
df_final_dataset['genres_cmu'] = df_final_dataset['genres_cmu'].apply(lambda x: str(x))
# Convertir les genres en listes
df_final_dataset['genres_cmu'] = df_final_dataset['genres_cmu'].apply(lambda x: x.split(', '))
# print the rows where countries are floats
print(df_final_dataset[df_final_dataset['countries'].apply(lambda x: isinstance(x, float))].head())
df_final_dataset['countries'] = df_final_dataset['countries'].apply(lambda x: x.split(', '))
# étendre le dataframe pour chaque genre
df_final_dataset = df_final_dataset.explode('genres_cmu')
df_final_dataset = df_final_dataset.explode('countries')
# Widget pour sélectionner le pays et le genre
country_selector = widgets.Dropdown(
    options=['All countries'] + list(df_final_dataset['countries'].unique()),  # Inclure 'All countries'
    description='Pays:',
)

genre_selector = widgets.Dropdown(
    options=['All genres'] + list(df_final_dataset['genres_cmu'].unique()),  # Inclure 'All genres'
    description='Genre:',
)

# Fonction pour mettre à jour le graphique en fonction du pays et du genre sélectionnés
def update_graph(country, genre):
    # Filtrer les données en fonction du pays sélectionné
    if country != 'All countries':
        filtered_data = df_final_dataset[df_final_dataset['countries'].notna() & df_final_dataset['countries'].str.contains(country)]
    else:
        filtered_data = df_final_dataset
    
    # Filtrer les données en fonction du genre sélectionné
    if genre != 'All genres':
        filtered_data = filtered_data[filtered_data['genres_cmu'].str.contains(genre)]
    
    # Limiter à 100 films maximum
    filtered_data = filtered_data.head(100)
    
    # Calculer le nombre de films pour chaque genre par pays
    genre_counts = filtered_data['genres_cmu'].value_counts()
    
    # Tracer les carrés
    total_movies = genre_counts.sum()
    num_squares = 100  # Nombre total de carrés à afficher
    squares_per_movie = num_squares / total_movies if total_movies != 0 else 0
    
    # Créer un tableau de carrés
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Position des carrés dans une grille 10x10
    for i in range(min(num_squares, total_movies)):
        row = i // 10
        col = i % 10
        
        genre = genre_counts.index[i % len(genre_counts)]  # Alterner les genres pour les carrés
        ax.add_patch(plt.Rectangle((col, row), 1, 1, color='lightblue' if genre == genre else 'lightgray'))

    # Afficher la grille
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_title(f"Films de genre {genre} et pays {country} (Total: {total_movies} films)")

    plt.tight_layout()
    plt.show()

# Lier les widgets à la fonction d'update
interactive_plot = interactive(update_graph, country=country_selector, genre=genre_selector)
display(interactive_plot)
```
